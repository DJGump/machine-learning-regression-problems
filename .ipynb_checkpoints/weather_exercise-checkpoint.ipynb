```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
bfriday_df = pd.read_csv("../weatherHistory.csv")
bfriday_df.head()
```

```python
# Y is the target variable
Y = bfriday_df['Apparent Temperature (C)'] - bfriday_df['Temperature (C)']

# X is the feature set
X = bfriday_df[['Humidity', 'Wind Speed (km/h)']]

# We add constant to the model as it's a best practice
# to do so everytime!
X = sm.add_constant(X)

# We fit an OLS model using statsmodels
results = sm.OLS(Y, X).fit()

# We print the summary results.
print(results.summary())
```

```python
plt.figure(figsize=(20,10))

plt.subplot(1,2,1)
plt.scatter(Y, bfriday_df[['Humidity']])

plt.subplot(1,2,2)
plt.scatter(Y, bfriday_df[['Wind Speed (km/h)']])
plt.show()
```

```python
# Y is the target variable
Y = bfriday_df['Apparent Temperature (C)'] - bfriday_df['Temperature (C)']

bfriday_df["Humidity_Wind_Speed"] = bfriday_df["Wind Speed (km/h)"]*bfriday_df["Humidity"]

# X is the feature set
X = bfriday_df[['Humidity', 'Wind Speed (km/h)', 'Humidity_Wind_Speed']]

# We add constant to the model as it's a best practice
# to do so everytime!
X = sm.add_constant(X)

# We fit an OLS model using statsmodels
results = sm.OLS(Y, X).fit()

# We print the summary results.
print(results.summary())
```
